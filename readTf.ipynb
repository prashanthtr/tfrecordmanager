```python
from tfrecordManager import tfrecordManager
```

#### Loads tfrecords for audio datasets in Patset.

```python
tfr = tfrecordManager()
```

```python
# Take a single record from a file
filename = ["sampleDir/shard0.tfrecord"]
data_record = tfr.__printRecord__(filename[0])
data_record
```

#### Load single records

```python
# Get feature list as JSON
tfr.__getFeatures__(data_record)
```

```python
# Use feature list to read serialized TFrecords
tfr.__readOne__(filename)
```

#### Loading tfrecords in a directory

```python
# Provide a relative path to read a directory of TFrecords
tfdir = "sampleDir"
tfr.__readDirectory__(tfdir)
```
