```python
import tensorflow as tf
import os
import argparse
import glob
```

#### Loads tfrecords for audio datasets in Patset.

```python
def printRecord(filename):

    raw_dataset = tf.data.TFRecordDataset(filename)
    print(raw_dataset)

    for raw_record in raw_dataset.take(1):
      tfToJSon = tf.train.Example()
      tfToJSon.ParseFromString(raw_record.numpy())
      print(tfToJSon)
```

#### Load single records

```python
filename = "sampleDir/sample.tfrecord"
printRecord(filename)
```

#### Loading tfrecords in a directory

```python
tfdir = "sampleDir"
path = os.path.abspath(tfdir)

if os.path.isdir(path):

    files = []
    for file in os.listdir(path):
        if file.endswith(".tfrecord"):
            printRecord(os.path.join(path, file))
else:
    print("Provide correct directory name")
```
